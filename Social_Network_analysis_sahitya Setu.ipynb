```python
#import libraries
import csv
from operator import itemgetter
import networkx as nx
from networkx.algorithms import community
import matplotlib.pyplot as plt
%matplotlib inline
%matplotlib notebook
import warnings; warnings.simplefilter('ignore')
```

```python
# File handling retriving data
with open(r'C:\Users\hp\Documents\social network analysis\Lastfm\lasftm_asia\lastfm_asia_target.csv', 'r') as nodecsv: 
    noderead = csv.reader(nodecsv) 
    nodes = [n for n in noderead][1:]

node_name_list = [n[0] for n in nodes] # node_id/name list

with open(r'C:\Users\hp\Documents\social network analysis\Lastfm\lasftm_asia\lastfm_asia_edges.csv', 'r') as edgecsv: 
    edgeread = csv.reader(edgecsv) 
    edges = [tuple(e) for e in edgeread][1:] # List of edges
```

```python
print('nodes  :' +str(len(node_name_list)) + '   '+'edges  :' + str(len(edges)))
```

```python
G_lastfm = nx.Graph()
G_lastfm.add_nodes_from(node_name_list)
G_lastfm.add_edges_from(edges)
print(nx.info(G_lastfm))
```

```python
plt.figure(figsize=(20,20))
nx.draw_networkx(G_lastfm)

plt.show()
```

```python
#between centrality
%matplotlib inline
pos = nx.spring_layout(G_lastfm)
betCent = nx.betweenness_centrality(G_lastfm, normalized=True, endpoints=True)
node_colour = [20000.0 * G_lastfm.degree(v) for v in G_lastfm]
node_sizes =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(G_lastfm, pos=pos, with_labels=True,
                 node_color=node_colour,
                 node_size=node_sizes )
plt.axis('off')
```

```python
# the top 20 nodes by betweenness as a list
betweenness_dict = nx.betweenness_centrality(G_lastfm)
sorted_betweenness = sorted(betweenness_dict.items(), key=itemgetter(1), reverse=True)
top_betweenness = sorted_betweenness[:20]
print(top_betweenness)
```

```python
nx.degree(G_lastfm)
```

```python
# the top 20 nodes by degree as a list
degree_dict = dict(G_lastfm.degree(G_lastfm.nodes()))

sorted_degree = sorted(degree_dict.items(), key=itemgetter(1), reverse=True)
top_degree = sorted_degree[:20]
print("Top 20 nodes by degree: " )
print(top_degree)
```

```python
# measures of distance 
print("diameter: ", nx.diameter(G_lastfm))

print("Eccentricity: ", nx.eccentricity(G_lastfm))

print("Periphery: ", nx.periphery(G_lastfm))

print("Center: ", nx.center(G_lastfm))
```
